## Relationship between 'New cases' and 'New deaths'

```python
plt.scatter(df.index, df['New cases'], alpha=0.5, color='blue', label='New Cases')

plt.scatter(df.index, df['New deaths'], alpha=0.5, color='red', label='New Deaths')

plt.title('Scatter Plot of New Cases vs New Deaths')
plt.xlabel('Index')
plt.ylabel('Count')
plt.legend()
plt.show()
```

## Proportion of 'Recovered' Cases for the Top 5 Countries using Pie Chart

```python
top_5_recovered = df.nlargest(5, 'Recovered')
plt.figure(figsize=(8, 8))
plt.pie(top_5_recovered['Recovered'], labels=top_5_recovered['Country/Region'], autopct='%1.1f%%', startangle=140)
plt.title('Proportion of Recovered Cases for Top 5 Countries')
plt.show()
```
