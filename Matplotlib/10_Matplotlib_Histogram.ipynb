## Histogram

#### A histogram is a graph showing frequency distributions. It is a graph showing the number of observations within each given interval.

## Create Histogram

#### In Matplotlib, we use the `hist()` function to create histograms.The `hist()` function will use an array of numbers to create a histogram, the array is sent into the function as an argument.

```python
# Normal Data Distribution
import numpy as np

x = np.random.normal(170, 10, 250)

print(x)
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.random.normal(170, 10, 250)
plt.hist(x)
plt.show() 
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Generate normally distributed numbers
mu, sigma = 0, 1  # mean and standard deviation
data = np.random.normal(mu, sigma, 1000)

# Create the histogram
plt.hist(data, bins=30, edgecolor='black')

# Add title and labels
plt.title('Histogram of Normally Distributed Data')
plt.xlabel('Value')
plt.ylabel('Frequency')

# Show the plot
plt.show()
```
